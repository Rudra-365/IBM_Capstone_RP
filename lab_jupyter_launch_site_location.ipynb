```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.plugins import MarkerCluster

# Initialize map
site_map = folium.Map(location=nasa_coordinate, zoom_start=4)

# Optional: use MarkerCluster to group nearby markers
marker_cluster = MarkerCluster().add_to(site_map)

# Iterate through each row in the dataframe and add a marker
for index, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    site = row['Launch Site']
    result = 'Success' if row['class'] == 1 else 'Failure'
    color = 'green' if row['class'] == 1 else 'red'

    # Add a colored circle marker for each launch
    folium.CircleMarker(
        location=coord,
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=folium.Popup(f"{site} - {result}", max_width=200)
    ).add_to(marker_cluster)

# Show the map
site_map
```

```python
from math import radians, sin, cos, sqrt, atan2
import pandas as pd

# Define haversine function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Confirm site names
print(launch_sites_df['Launch Site'].unique())

# Correct dictionary of nearby features
landmarks = {
    'CCAFS LC-40': (28.3200, -80.6076),    # Cocoa Beach, FL
    'VAFB SLC-4E': (34.6390, -120.4579),   # Lompoc, CA
    'KSC LC-39A': (28.6122, -80.8076),     # Titusville, FL
    'CCAFS SLC-41': (28.3922, -80.6077)    # Cape Canaveral
}

# Calculate distances
launch_sites_df['Distance (km)'] = launch_sites_df.apply(
    lambda row: round(
        calculate_distance(
            row['Lat'],
            row['Long'],
            landmarks[row['Launch Site']][0],
            landmarks[row['Launch Site']][1]
        ),
        2
    ),
    axis=1
)

# Display result
launch_sites_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
